```python
# Figure 3.1 Average housing price time series plot (London and Northwest)

import pandas as pd
import matplotlib.pyplot as plt


london_df = pd.read_csv('London_Housing_Policy_Dataset.csv')
north_west_df = pd.read_csv('North_West_Housing_Policy_Dataset.csv')

london_df['Date'] = pd.to_datetime(london_df['Date'])
london_df.set_index('Date', inplace=True)

north_west_df['Date'] = pd.to_datetime(north_west_df['Date'])
north_west_df.set_index('Date', inplace=True)

plt.figure(figsize=(12, 4))
plt.plot(london_df.index, london_df['Average_Housing_Price'], label='London', color='tab:blue')
plt.plot(north_west_df.index, north_west_df['Average_Housing_Price'], label='North West', color='tab:orange')
plt.title('Average Housing Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```

```python
# Figure 3.2 Policy variables time series plot (London and Northwest)

import pandas as pd
import matplotlib.pyplot as plt

london_df = pd.read_csv('London_Housing_Policy_Dataset.csv')
north_west_df = pd.read_csv('North_West_Housing_Policy_Dataset.csv')

london_df['Date'] = pd.to_datetime(london_df['Date'])
london_df.set_index('Date', inplace=True)

north_west_df['Date'] = pd.to_datetime(north_west_df['Date'])
north_west_df.set_index('Date', inplace=True)


variables = [
    'Weekly_Earnings',
    'Employment_Rate',
    'Benefit Caseload',
    'Housing Supply',
    'Euro Rate',
    'Interest Rate'
]

# 3x2
fig, axes = plt.subplots(3, 2, figsize=(10, 6), sharex=False)

for i, (ax, var) in enumerate(zip(axes.flat, variables)):
    if var in ['Euro Rate', 'Interest Rate']:
        # 只画伦敦
        ax.plot(london_df.index, london_df[var], color='tab:blue')
    else:
        ax.plot(london_df.index, london_df[var], label='London', color='tab:blue')
        ax.plot(north_west_df.index, north_west_df[var], label='North West', color='tab:orange')
        ax.legend()
    ax.set_title(var.replace('_', ' '))
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Figure 3.3 Pair plot of London housing policy data

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("London_Housing_Policy_Dataset_sd.csv", parse_dates=["Date"])

price_series = df["Average_Housing_Price"].values
policy_columns = ['Weekly_Earnings','Employment_Rate','Benefit Caseload','Housing Supply','Euro Rate','Interest Rate']

# Combine housing price and policies
columns_to_plot = ['Average_Housing_Price'] + policy_columns
df_plot = df[columns_to_plot]
sns.set(style="ticks", font_scale=1.2)
sns.pairplot(df_plot)
plt.show()
sns.reset_orig()
```

```python
# Figure 3.4 Pair plot of Northwest housing policy data(Part)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("North_West_Housing_Policy_Dataset_sd.csv", parse_dates=["Date"])

policy_columns = ['Weekly_Earnings','Employment_Rate','Benefit Caseload','Housing Supply','Euro Rate','Interest Rate']
columns_to_plot = policy_columns  

df_plot = df[['Average_Housing_Price'] + policy_columns]

sns.set(style="ticks", font_scale=1.2)

g = sns.PairGrid(df_plot, x_vars=columns_to_plot, y_vars=['Average_Housing_Price'], height=3)
g.map(sns.scatterplot, color="tab:orange")

plt.show()
sns.reset_orig()
```
