```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
london_data = pd.read_csv('London_Housing_Policy_Dataset.csv', parse_dates=['Date'])
london_data = london_data.sort_values('Date')

scaler = StandardScaler()
X_scaled = scaler.fit_transform(london_data.drop(columns=['Date']))
X = pd.DataFrame(X_scaled, columns=london_data.columns[1:], index=london_data['Date'])
london_data_sd=X.reset_index()
london_data_sd.to_csv("London_Housing_Policy_Dataset_sd.csv", index=False)
```
